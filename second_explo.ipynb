```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

df = pd.read_csv("/Users/margauxelbaum/Desktop/BossLadies/archive/cleaned_ingredients.csv")

# ── 1. NaN values ──────────────────────────────────────────────────────────────
print("=" * 60)
print("1. NaN VALUES")
print("=" * 60)
nan_counts = df.isnull().sum()
nan_pct = (df.isnull().mean() * 100).round(2)
nan_df = pd.DataFrame({"NaN count": nan_counts, "NaN %": nan_pct})
nan_df = nan_df[nan_df["NaN count"] > 0].sort_values("NaN count", ascending=False)
if nan_df.empty:
    print("No NaN values found!")
else:
    print(nan_df.to_string())
print(f"\nTotal NaN cells: {df.isnull().sum().sum()}")

# ── 2. Min / Max per numeric column ────────────────────────────────────────────
print("\n" + "=" * 60)
print("2. MIN / MAX PER COLUMN")
print("=" * 60)
numeric_cols = df.select_dtypes(include=[np.number]).columns
minmax = pd.DataFrame({
    "min": df[numeric_cols].min(),
    "max": df[numeric_cols].max()
})
print(minmax.to_string())

# ── 3. Non-feasible values ─────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("3. NON-FEASIBLE VALUES")
print("=" * 60)

# 3a. Energy > 900 kcal
if "Energy_kcal" in df.columns:
    mask_kcal = df["Energy_kcal"] > 900
    print(f"\n[Energy > 900 kcal]  {mask_kcal.sum()} rows")
    if mask_kcal.sum():
        cols_show = ["NDB_No", "Descrip", "Energy_kcal"] if "Descrip" in df.columns else ["NDB_No", "Energy_kcal"]
        print(df.loc[mask_kcal, [c for c in cols_show if c in df.columns]].to_string(index=False))

# 3b. g-per-100g columns that exceed 100
over100_cols = ["Protein_g", "Fat_g", "Saturated_fats_g", "Carb_g", "Fiber_g", "Sugar_g"]
over100_cols = [c for c in over100_cols if c in df.columns]

for col in over100_cols:
    mask = df[col] > 100
    print(f"\n[{col} > 100 g]  {mask.sum()} rows")
    if mask.sum():
        id_cols = [c for c in ["NDB_No", "Descrip"] if c in df.columns]
        print(df.loc[mask, id_cols + [col]].to_string(index=False))

# 3c. Any negative values in numeric columns
print("\n[Negative values in numeric columns]")
neg_found = False
for col in numeric_cols:
    neg_mask = df[col] < 0
    if neg_mask.sum() > 0:
        neg_found = True
        print(f"  {col}: {neg_mask.sum()} negative rows")
if not neg_found:
    print("  None found.")

print("\n" + "=" * 60)
print("Done.")
```

We can see that all are capped at max 100g for the important columns, now I want to see whether the sum all macros doesn't exceed 100g (knowing that both fiber and sugar are a subset of carbs)

```python
# Correct macro sum: no double-counting
g_cols_correct = ["Protein_g", "Fat_g", "Carb_g"]
df["macro_sum_correct"] = df[g_cols_correct].sum(axis=1)
mask_correct = df["macro_sum_correct"] > 100

print(f"[Protein + Fat + Carb only] Rows > 100g: {mask_correct.sum()}")

if mask_correct.sum() > 0:
    id_cols = [c for c in ["NDB_No", "Descrip"] if c in df.columns]
    print(df.loc[mask_correct, id_cols + g_cols_correct + ["macro_sum_correct"]].to_string(index=False))
```

We can see that there are only 2 rows that are not exactly capped at 100, and those are only a little above 100g. Most the values are under 100g, but this is probably due to the water in them

Next thing we need to know, is how many food items are at 0.0 for one of the macros, this could either mean those values are truly zeros, or they are NaN, and we need to handle them 

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if "NDB_No" in numeric_cols:
    numeric_cols.remove("NDB_No")
numeric_cols = pd.Index(numeric_cols)

print(numeric_cols)  # verify before proceeding

zero_counts = (df[numeric_cols] == 0.0).sum()
zero_pct = ((df[numeric_cols] == 0.0).mean() * 100).round(2)

zero_df = pd.DataFrame({
    "zero count": zero_counts,
    "zero %": zero_pct,
    "NaN count": df[numeric_cols].isnull().sum(),
    "NaN %": (df[numeric_cols].isnull().mean() * 100).round(2)
}).sort_values("zero %", ascending=False)

print("Zeros vs NaNs per column:")
print(zero_df.to_string())
```

```python
check_cols = ["Energy_kcal", "Protein_g", "Fat_g", "Carb_g", "Fiber_g", "Sugar_g"]

for col in check_cols:
    mask_zero = df[col] == 0.0
    n_zeros = mask_zero.sum()
    print(f"\n{'='*60}")
    print(f"{col} — {n_zeros} zeros ({n_zeros/len(df)*100:.1f}%)")
    if n_zeros > 0:
        print(df.loc[mask_zero, ["Descrip"] + check_cols].to_string(index=False))

# ── Rows where ALL three main macros are 0 ─────────────────────────────────────
print(f"\n{'='*60}")
mask_all_zero = (df["Protein_g"] == 0.0) & (df["Fat_g"] == 0.0) & (df["Carb_g"] == 0.0)
print(f"Rows where Protein + Fat + Carb are ALL 0.0: {mask_all_zero.sum()}")
if mask_all_zero.sum() > 0:
    print(df.loc[mask_all_zero, ["Descrip"] + check_cols].to_string(index=False))
```

I also wanted to see, which food items had their sum at 0.0, to see if those were missing values or true. 
All of those are either pure minerals, water (different brands), zero sports drinks, zero calories beverages or seasonix mix. Also, those 39 rows correspond to the number of rows that kcal also has as 0.  

```python
check_cols = ["Energy_kcal", "Protein_g", "Fat_g", "Carb_g", "Fiber_g", "Sugar_g"]

# collect all rows with at least one zero in macro cols
mask_any_zero = (df[check_cols] == 0.0).any(axis=1)
df_zero_items = df.loc[mask_any_zero, ["NDB_No", "Descrip"] + check_cols].copy()

# add a column showing WHICH macros are zero for each row
df_zero_items["zero_in"] = df.loc[mask_any_zero, check_cols].apply(
    lambda row: ", ".join([col for col in check_cols if row[col] == 0.0]), axis=1
)

print(f"Total rows with at least one macro at 0: {len(df_zero_items)}")

# save to csv
df_zero_items.to_csv("/Users/margauxelbaum/Desktop/BossLadies/zero_macro_items.csv", index=False)
print("Saved to /Users/margauxelbaum/Desktop/BossLadies/zero_macro_items.csv")
```

After checking the csv, I think most of the values are correct, particularly for fiber, as most of the missing values in the csv file are fiber, but lots of items don't contain any fiber. However, one column that needs a bit more attention is sugar. As some items are highly probable to have sugar in them, such as muffin, donuts, pie, etc. However, most of the values for sugar are still correct, as many columns include different types of meat. 

```python
suspicious_keywords = ["muffin", "donut", "doughnut", "pie", "cake", "cookie", 
                       "brownie", "pastry", "waffle", "pancake", "syrup", "candy",
                       "chocolate", "ice cream", "pudding", "tart", "croissant", "granola bar"]

pattern = "|".join(suspicious_keywords)

mask_suspicious_sugar = (
    (df["Sugar_g"] == 0.0) & 
    (df["Descrip"].str.contains(pattern, case=False, na=False))
)

print(f"Suspicious sugar zeros: {mask_suspicious_sugar.sum()}")
print(df.loc[mask_suspicious_sugar, ["Descrip", "Energy_kcal", "Carb_g", "Sugar_g"]].to_string(index=False))

df["sugar_unreliable"] = mask_suspicious_sugar
```

Some of those values are definitely suspicious, such as syrup, which is mainly sugar, however "ex lt" means extra light, but they should still contain some sugar. 
One option would be to flag those, don't rely on them, except if there is a "sugar free" indication?

```python
suspicious_keywords = [
    "muffin", "donut", "doughnut", "pie", "cake", "cookie",
    "brownie", "pastry", "waffle", "pancake", "candy",
    "chocolate", "ice cream", "pudding", "tart", "croissant",
    "syrup", "agutuk", "arrowroot babyfood"
]

pattern = "|".join(suspicious_keywords)

mask_suspicious_sugar = (
    (df["Sugar_g"] == 0.0) &
    (df["Descrip"].str.contains(pattern, case=False, na=False)) &
    (~df["Descrip"].str.contains("sugar free|sugarfree|no sugar|unsweetened", case=False, na=False))
)

df["sugar_unreliable"] = mask_suspicious_sugar

print(f"Flagged as sugar_unreliable: {mask_suspicious_sugar.sum()} rows")
print(df.loc[mask_suspicious_sugar, ["Descrip", "Carb_g", "Sugar_g"]].to_string(index=False))
```

Out of the 181 values, 14 are unsweetened, so truly 0 sugar. 

Overall, this is a very good dataset, much more reliable than the other one.
